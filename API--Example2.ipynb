```python
import pandas as pd
URL = "https://data.cityofnewyork.us/resource/k397-673e.json"
```

```python
import requests
```

```python
dfs = []
for i in range(100):
    payload = {'$limit': 50000,
              '$offset': 50000 * i}
    resp = requests.get(url = URL, params = payload)
    assert resp.status_code == 200
    df = pd.DataFrame(resp.json())
    dfs.append(df)
```

```python
bdf = pd.concat(dfs)
```

```python
bdf.columns
```

```python
dat = pd.DataFrame(bdf)
```

```python
dat.head()
```

```python
# between 2019 and 2020, which month had the most people starting in their agency?
import datetime as dt
def get_year(time):
    if time != 'nan':
        time_new = dt.datetime.strptime(time, '%Y-%m-%dT%H:%M:%S.000')
    else:
        time_new = ''
    return time_new
year = dat.agency_start_date.apply(lambda x: get_year(str(x)))
```

```python
dat.head()
```

```python
def get_year2(time):
    if isinstance(time, str):
        year = time[: 4]
    else:
        year = ''
    return year
year_2 = bdf.agency_start_date.apply(get_year2)
```

```python
is_2019 = year_2 == "2019"
is_2020 = year_2 == "2020"
is_2019or2020 = is_2019 | is_2020
```

```python
is_2019or2020
```

```python
dat.shape
```

```python
sdf = dat.loc[is_2019or2020, :].copy()
```

```python
sdf.head()
```

```python
sdf['month'] = sdf.agency_start_date.apply(lambda x: x[5: 7])
```

```python
sdf.groupby('month').count().agency_name
```

```python
sdf.groupby('month').count().agency_name.argmax()
```

```python
month_cnt = pd.DataFrame(sdf.groupby('month').count().agency_name).reset_index()
```

```python
month_cnt
```

```python
max_month = month_cnt.month[month_cnt.agency_name.argmax()]
```

```python
print(f"The month with the most people starting is {max_month}")
```

```python
# hwo to get the columns in a data which has only one value!
sin_col = [col for col in df.columns
          if df[col].drop_duplicates().shape[0] == 1]
```

```python
df.drop(columns=sin_col, inplace=True)
```

```python
sdf.Value = sdf.Value.apply(lambda x: float(x.replace(',', '')) if x.find(' ') < 0 else np.nan) 
```

```python
sdf.head()
```

```python
acres_df = sdf.loc[sdf.unit_desc == "ACRES"]
```

```python
acres_df.shape
```

```python
bushel_df = sdf.loc[sdf.unit_desc == "BU"]
```

```python
bushel_df.shape
```

```python
dedup_acres_df = acres_df.groupby(["state_name", "year"], as_index=False).max()
```

```python
dedup_acres_df
```

```python
dedup_bushel_df = bushel_df.groupby(["state_name", "year"], as_index=False).max()
```

```python
dedup_bushel_df
```

```python
# we can see when we don't apply method into the groupby result, what can we get--
grp_df = bushel_df.groupby(["state_name", "year"], as_index=False)
```

```python
group = ('ALABAMA', 1997)
```

```python
# we can use the get_group() method to get the group in the groupby result.
```

```python
grp_df.get_group(group)
```

```python
jdf = dedup_acres_df.merge(dedup_bushel_df, on = ['state_name', 'year'])
```

```python
jdf.head()
```

```python
jdf["yield"] = jdf.Value_y / jdf.Value_x
```

```python
jdf.head()
```

```python
# Task 2 - Wrangle the data
# Please wrangle the data so we can plot the change in yields over the years for each state.
```

```python
sns.relplot(x = "year", y = "yield", hue="state_name", data=jdf, kind="line")
```

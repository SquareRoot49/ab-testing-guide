```python
max_bushel_df = bushel_df.groupby(['state_name', 'year'], as_index = False).max()
```

```python
max_acres_df = acres_df.groupby(['state_name', 'year'], as_index = False).max()
```

```python
jdf = max_acres_df.merge(max_bushel_df, on  = ['state_name', 'year'])
```

```python
jdf.rename(columns = {'Value_x': 'acres', 'Value_y': 'bushels'}, inplace = True)
```

```python
jdf.acres = jdf.acres.astype(float)
```

```python
jdf.bushels = jdf.bushels.astype(float)
```

```python
jdf["yield"] = jdf.bushels / jdf.acres
```
